```python
import pandas as pd 
import numpy as np 

df = pd.read_excel('official2.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
df_alpha_phone = df[df['phone'].str[0].str.isalpha()]
print(len(df_alpha_phone))
display(df_alpha_phone)
```

```python
from selenium import webdriver 
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By

phone_list = []
count = 0 

for url in df_alpha_phone['link']:
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)
    page = driver.page_source 
    soup = BeautifulSoup(page, 'html.parser')

    try:
        show_phone_number_button = driver.find_element(By.ID, 'showPhoneNumber')
        show_phone_number_button.click()

        # Wait for the phone number to be displayed
        time.sleep(2)

        # Get the updated page source
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        phone_number_span = soup.find('span', {'id': 'ctl00_ctl00_Content_ContentPlaceHolder1_ctl02_BrokerTopInfo_lblTelephone'})
        phone_number = phone_number_span.get_text(strip=True) if phone_number_span else 'Phone number not found'
    except:
        print(count, 'N/A')
        count += 1 
        phone_list.append(np.nan)
    else:
        print(count, phone_number)
        count += 1 
        phone_list.append(phone_number)
```

```python
del df_alpha_phone['phone']
df_alpha_phone.head()
```

```python
df_alpha_phone['phone'] = phone_list
df_alpha_phone.head()
```

```python
df_cleaned = df[~df['phone'].str[0].str.isalpha()]
df_cleaned.shape
```

```python
df2 = pd.concat([df_cleaned, df_alpha_phone], ignore_index = True)
df2.head()
```
