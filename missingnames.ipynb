```python
import pandas as pd
import numpy as np

df = pd.read_excel('bizbuysell.xlsx')
df.head()
```

```python
del df['Unnamed: 0']
df.head()
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time 

# Name
name_null = df[df['name'].isnull()]
display(name_null)
```

```python
name_null_names = []
count = 0

for url in name_null['link']:
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(10)
    page = driver.page_source 
    soup = BeautifulSoup(page, 'html.parser')

    try:
        location = soup.find('div', {'class': 'width-fit remove-padding normal f-l'})
        name = location.get_text(strip=True)
    except:
        print(count, 'N/A')
        count += 1
        name_null_names.append(np.nan)
    else:
        print(count, name)
        count += 1
        name_null_names.append(name)

print(name_null_names)
```

```python
del name_null['name']
name_null.head()
```

```python
name_null['name'] = name_null_names
name_null.head()
```

```python
new_order = ['name', 'phone', 'company', 'state', 'link']
name_null = name_null[new_order]
name_null.head()
```

```python
df = df.dropna(subset=['name'])
df = pd.concat([df, name_null], ignore_index=True)
df.head()
```

```python
print(len(df))
```

```python
df.to_excel('fixed_name.xlsx')
```
