```python
import pandas as pd 
import numpy as np 

df = pd.read_excel('official.xlsx')
del df['Unnamed: 0']
df.shape
df.head()
```

```python
df.shape
```

```python
name_list = []
for name in df['name']:
    if ',' in name:
        new = name[:name.index(',')]
        name_list.append(new)
        print(new)
    else:
        name_list.append(name)
        print(name)
```

```python
print(len(name_list))
```

```python
del df['name']
df.head()
```

```python
df['name'] = name_list 
df.head()
```

```python
df.rename(columns={'state': 'location'}, inplace = True)
df.head()
```

```python
state_list = []
city_list = []

for location in df['location']:
    if ',' in location:
        division = location.split(',')
        print(division)
        city_list.append(division[0])
        state_list.append(division[1])
        
    else:
        print(location, 'cannot be split')
        state_list.append(location)
        city_list.append(np.nan)

print(len(state_list))
print(len(city_list))
```

```python
df['state'] = state_list 
df['city'] = city_list 
df.head()
```

```python
for name in df['name']:
    divi = name.split(' ')
    if len(divi) == 1:
        print(name)
```

```python
first = []
last = []

for name in df['name']:
    divide = name.split(' ')
    print(divide)
    
    if len(divide) == 2:
        first.append(divide[0])
        last.append(divide[-1])

    elif len(divide) == 3:
        first_name = divide[0] + ' ' + divide[1]
        first.append(first_name)
        last.append(divide[-1])
    
    else:
        first.append(np.nan)
        last.append(np.nan)
```

```python
print(first)
print(last)
```

```python
df['first name'] = first 
df['last name'] = last 
df.head()
```

```python
df = df.sort_values(by='state')
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df.to_excel('official2.xlsx')
```
