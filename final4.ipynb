```python
import pandas as pd
import numpy as np 

df = pd.read_excel('official4.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
for num in range(len(df['Corporate Phone'])):
    phone = df.loc[num, 'Corporate Phone']
    
    if pd.notnull(phone):
        phone = str(phone)
        new = ''
        for character in phone:
            if character.isnumeric() == True:
                new += character 
        phone = '+1 ' + new[:3] + '-' + new[3:6] + '-' + new[6:]
        df.loc[num, 'Corporate Phone'] = phone
```
