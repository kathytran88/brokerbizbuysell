```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel('fixed_name.xlsx')
df.head()
```

```python
del df['Unnamed: 0']
df.head()
```

```python
company_null = df[df['company'].isnull()]
company_null.head()
```

```python
print(len(company_null))
```

```python
from selenium import webdriver 
from bs4 import BeautifulSoup
import time

company_null_list = []
count = 0

for url in company_null['link']:
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(10)
    page = driver.page_source 
    soup = BeautifulSoup(page, 'html.parser')

    try:
        location = soup.find('span', {'class': "f-m margin-b-8 block normal"})
        company = location.get_text(strip=True)
    except:
        print(count, 'N/A')
        count += 1
        company_null_list.append(np.nan)
    else:
        print(count, company)
        count += 1
        company_null_list.append(company)

print(company_null_list) 
```

```python
print(company_null_list)
print(len(company_null_list))
```

```python
for a in range(32):
    company_null_list.append(np.nan)

print(len(company_null_list))
```

```python
del company_null['company']
company_null.head()
```

```python
company_null['company'] = company_null_list
company_null.head()
```

```python
df = df.dropna(subset= ['company'])
df = pd.concat([df, company_null], ignore_index=True)
df.head()
```

```python
print(len(df))
```

```python
df.to_excel('fixed_company.xlsx')
```
