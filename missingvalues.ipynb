```python
import pandas as pd

df = pd.read_excel('bizbuysell.xlsx')
df.head()
```

```python
del df['Unnamed: 0']
df.head()
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time 

# Name
name_null = df[df['name'].isnull()]
display(name_null)
```

```python
name_null_names = []
for url in name_null['link']:
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(10)
    page = driver.page_source 
    soup = BeautifulSoup(page, 'html.parser')

    location = soup.find('div', {'class': 'width-fit remove-padding normal f-l'})
    name = location.get_text(strip=True)
    name_null_names.append(name)

print(name_null_names)
```
