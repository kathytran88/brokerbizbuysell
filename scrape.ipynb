```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome() 

driver.get('https://www.bizbuysell.com/business-brokers/wyoming/')

time.sleep(5)  

page_source = driver.page_source

soup = BeautifulSoup(page_source, 'html.parser')

# Elite
broker_cards = soup.find_all('app-bfs-elite-brokercard-search-result') 

urls = []
for card in broker_cards:
    link = card.find('a', href=True)
    if link:
        link = 'https://www.bizbuysell.com' + link['href']
        urls.append(link)

# Premium
broker_cards2 = soup.find_all('app-bfs-premium-brokercard-search-result')
if broker_cards2:
    for card in broker_cards2:
        link = card.find('a', href=True)
        if link:
            link = 'https://www.bizbuysell.com' + link['href']
            urls.append(link)
# Basic
broker_cards3 = soup.find_all('app-bfs-basic-brokercard-search-result')
if broker_cards3:
    for card in broker_cards3:
        link = card.find('a', href=True)
        if link:
            link = 'https://www.bizbuysell.com' + link['href']
            urls.append(link)

# Print
for url in urls:
    print(url)

driver.quit()

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()

profiles = []

try:
    for url in urls:
        driver.get(url)
        driver.implicitly_wait(10)  

        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')

        # Dictionary to store profile information
        profile = {
            'name': 'N/A',
            'phone': 'N/A',
            'company': 'N/A',
            'state': 'N/A',
            'link': 'N/A'
        }
        
        profile['link'] = url

        # Name - Handle different possible locations
        name_location = soup.find('div', class_='width-fit remove-padding normal f-l')
        if name_location:
            profile['name'] = name_location.get_text(strip=True)
        else:
            name_location1 = soup.find("span", {"id": "ctl00_ctl00_Content_ContentPlaceHolder1_ctl02_BrokerTopInfo_lblName"})
            if name_location1:
                profile['name'] = name_location1.get_text(strip=True)
            else:
                # Debugging: Print the entire HTML content to inspect
                print(f"Name location not found in {url}")

        # Phone number 
        try:
            show_phone_number_button = driver.find_element(By.ID, 'showPhoneNumber')
            show_phone_number_button.click()

            # Wait for the phone number to be displayed
            time.sleep(2)

            # Get the updated page source
            page_source = driver.page_source
            soup = BeautifulSoup(page_source, 'html.parser')

            phone_number_span = soup.find('span', {'id': 'ctl00_ctl00_Content_ContentPlaceHolder1_ctl02_BrokerTopInfo_lblTelephone'})
            phone_number = phone_number_span.get_text(strip=True) if phone_number_span else 'Phone number not found'
            profile['phone'] = phone_number
        except Exception as e:
            print(f"Error extracting phone number for {url}: {e}")

        # Company
        company_location = soup.find('span', class_='f-m margin-b-8 block normal')
        if company_location:
            profile['company'] = company_location.get_text(strip=True)

        # State 
        state_location = soup.find('p', class_='business-location')
        if state_location:
            profile['state'] = state_location.get_text(strip=True)
        else:
            state_location1 = soup.find('span', id="ctl00_ctl00_Content_ContentPlaceHolder1_ctl02_ctl00_lblAreasServed")
            if state_location1:
                profile['state'] = state_location1.get_text(strip=True)

        # Add profile to profiles list
        profiles.append(profile)

        # Print profile information
        print(f"Name: {profile['name']}, Phone: {profile['phone']}, Company: {profile['company']}, State: {profile['state']}")

finally:
    driver.quit()
```
