```python
import pandas as pd 
import numpy as np 

df = pd.read_excel('official.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df_unique = df.drop_duplicates(keep='first')
print(len(df_unique))
```

```python
df_unique.to_excel('unique1.xlsx')
```

```python
name_duplicates_mask = df_unique['name'].duplicated(keep=False)
duplicate_names_df = df_unique[name_duplicates_mask]

display(duplicate_names_df)
```

```python
name_dup = df_unique['name'].duplicated(keep=False)
name_dup_df = df_unique[name_dup]
display(name_dup_df)
```

```python
df_sort = df_unique.sort_values(by = 'name')
display(df_sort)
```

```python
df_sort = df_sort.reset_index(drop=True)
df_sort.head()
```

```python
df_sort.to_excel('official2.xlsx')
```
