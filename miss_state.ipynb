```python
import pandas as pd 
import numpy as np 

df = pd.read_excel('official1.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
state_null = df[df['state'].isnull()]
print(len(state_null))
display(state_null)
```

```python
from selenium import webdriver 
from bs4 import BeautifulSoup 
import time 

state_null_list = []
count = 0 

for url in state_null['link']:
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)
    page = driver.page_source 
    soup = BeautifulSoup(page, 'html.parser')

    try:
        location = soup.find('span', {'id': 'ctl00_ctl00_Content_ContentPlaceHolder1_ctl02_ctl00_lblAreasServed'})
        state = location.get_text(strip=True)
    except:
        print(count, 'N/A')
        count += 1 
        state_null_list.append(np.nan)
    else:
        print(count, state)
        count += 1 
        state_null_list.append(state)
```

```python
del state_null['state']
state_null.head()
```

```python
state_null['state'] = state_null_list 
state_null.head()
```

```python
print(len(df))
```

```python
print(len(state_null))
```

```python
df = df.dropna(subset=['state'])
print(len(df))
```

```python
df = pd.concat([df, state_null], ignore_index=True)
print(len(df))
```

```python
df.to_excel('official2.xlsx')
```
